```python
## Let's learn plotting

# 1st step
import matplotlib.pyplot as plt

y = [10, 20, 10, 40]
x = [1,2,3,4]

plt.plot(x,y)
plt.show()
```

```python
# make the plot better looking with some labels
plt.plot(x,y)
plt.xlabel("identifying number")
plt.ylabel("corresponding values")
plt.show()
```

```python
# add grid
plt.plot(x,y)
plt.xlabel("identifying number")
plt.ylabel("corresponding values")
plt.grid(True) # this line
plt.show()
```

```python
# add title
plt.plot(x,y)
plt.xlabel("identifying number")
plt.ylabel("corresponding values")
plt.grid(True)
plt.title("This is not a meaningful title")# this line
plt.show()
```

```python
# add markers
# check this link for more options: https://matplotlib.org/3.1.1/api/markers_api.html
plt.plot(x,y, marker='o') # this line
plt.xlabel("identifying number")
plt.ylabel("corresponding values")
plt.grid(True)
plt.title("This is not a meaningful title")
plt.show()
```

```python
# add amrker, change linewidth
# check this link for more options: https://matplotlib.org/3.1.1/api/markers_api.html
plt.plot(x,y,'g*-', linewidth=15) # this line
plt.xlabel("identifying number")
plt.ylabel("corresponding values")
plt.grid(True)
plt.title("This is not a meaningful title")
plt.show()
```

```python
# plot a sin curve
import math
import numpy as np
x = np.arange(-math.pi, math.pi, 0.5).tolist()
y = [math.sin(i) for i in x]

plt.plot(x, y,'go-.', linewidth=1, markersize=5, markerfacecolor='b') # this line
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.title("Plot of sin(x)")
plt.show()
```
