First, what are loops?

Think of it running in circles. You start at a point, run for a while, and come back to the same point. But when it comes to programming, we do a bit more.

You might ask, when do we stop? If we don't know when to stop, we might keep running endlessly (*that may actually happen!*). 

In programming, when we are running in circles, and check if a condition is true.

Think of it this way:

    * while the temperature is cold,
        wear a jacket
        
    * while the number of items in the shopping list is not zero,
        keep shopping

So we know a condition that determines our actions, but not necesarily how long we will keep doing that stuff.





* While loop typically reads in first value before loop – Initial Read

* Each subsequent read is called an Update Read or Subsequent Read
        
   

```python
# Let's say we want to print Hello World! five times

count = 2

while (count <= 5):
    """
    This function counts until 5 given a start number that is less
    than 5. 
    """
    # This function counts until 5 given a start number that is less
    # than 5. 
    print("Hello World!")
    print(count)
    count = count + 1 # same as count = count + 1;
    # count += 1

    
    
```

Notes: 
* Note that we had to initialize the loop before the while loop block started
* There is a : after we have stated the condition
* After while loop condition has been specified, there is an indentation before the next line starts
* We had to update *count* inside the while loop
* Q: What would happen if we didn't update "count" within the loop?


```python
count = 0

while count < 5:
    print("Hello World!")
```

```python
flag = 0

while (flag == 0):
    data = str(input("Enter text:\n"))
    if data == 'exit':
        flag = 1

```

* We must update the variable that is used in the condition for the example above

```python
count = 5

while count > 0:
    print("Hello World!")
    count += 1   # same as count = count + 1
```

What's wrong this time?

* We must make sure that updating the condition will make the condition **false** eventually 

Two typical ways to implement while loop conditions:
* flag
* counter


```python
num_list = [3,4,2,5,56,7,8]

sum = 0
flag = 0
index = 0
iteration = 0

while flag == 0:
    sum += num_list[index] # sum = sum + num_list[0]  , for the 0th iteration; sum = sum + num_list[1], for the 1st iteration
    if sum > 20:
        flag = 1
    
    
    print("iteration")
    print(iteration)
    print("flag")
    print(flag)
    print("index")
    print(index)
    print("current item from num_list")
    print(num_list[index])
    print("sum")
    print(sum)
    print("\n\n")
    
    index = index + 1 # index += 1
    iteration += 1
    
print("DONE")    
        

```

```python
# break statement

count = 1

while count < 500:
    #print("hello world!")
    print(count)
    if count == 10:
        break
    count += 2
    

```

```python
# continue

count = 0

while count < 7:
    print("hello world!")
    
    count += 1
    if count == 3:
        continue
    print(count)
    
```
