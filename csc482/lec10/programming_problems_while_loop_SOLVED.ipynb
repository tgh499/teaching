### Problem 1
Generate as many random numbers between 0 and 5 as needed to get a sum less than or equal to 50. The moment the sum crosses 50, we stop counting. Display how many random numbers you generated. Make sure to use a while loop. Search on Google how to generate random numbers.

```python
import random

sum = 0
count = 0

while (sum <= 50):
    sum += random.randint(0,5) # sum = sum + random.randint(0,5)
    count += 1 # count = count + 1
    
print(sum)
print(count)
```

### Problem 2
Write a program using a ***while loop*** that finds the values less than or equal to 80 from ***rand_arr***, adds them all, and then displays the sum. 

```python
# DON"T CHANGE THE CODE HERE
import numpy as np
rand_arr = [[10, 18, 12, 8, 34], [1, 8, 115, 9, 123], [35, 12, 921, 13, 56], [25, 0, 16, 5, 99]]
print("This is how rand_arr looks like in 2D: \n")
print(np.matrix(rand_arr))
```

```python
# TYPE YOUR CODE IN THIS CELL

row = 0
col = 0
sum = 0

row_length = len(rand_arr)
col_length = len(rand_arr[0])

while row < row_length:
    col = 0  # why do you need to reset col to 0?
    while col < col_length:
        if rand_arr[row][col] <= 80:
            sum += rand_arr[row][col]
            #print(row)
            #print(col)
            #print(rand_arr[row][col])
        col += 1
    row += 1
    
    
print(sum)
```

### Problem 3

Write program to generate a pattern like the following. Use while loop.



```python
row_len = 17
col_len = 13

row = 0
col = 0

while row < row_len:
    col = 0
    while col < col_len:
        if row % 2 != 0 and col % 2 != 0:
            print("1", end = '')
        else:
            print("*", end = '')
        col += 1
    row += 1
    print("\n")
```
