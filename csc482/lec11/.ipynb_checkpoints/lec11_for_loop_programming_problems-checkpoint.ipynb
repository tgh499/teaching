1. Write a program that finds the sum of 30 numbers that increments by 4. Your starting number should be 33.

```python
temp = 33
sum = 0

for i in range(30):
    print(i)
    print(temp)
    
    sum += temp
    temp += 4
    
    print(sum)
    print("\n\n")
    

```

2. Wirte a program to count the number of vowels in a given sentence.


```python
text = "Hello Universe!!"
text = text.lower()

print(text)
vowel = 0 

for i in range(len(text)):
    if (text[i] == 'a' or text[i] == 'e' or text[i] == 'i' 
                    or text[i] == 'o' or text[i] == 'u'):
        
        vowel += 1

print(vowel)


```

3. You have been given the following array. 

```python
import numpy as np

A = [[1,2,3],[4,5,6],[7,8,9]]

B  = np.zeros((len(A), (len(A)+2)))

#print(A)


for row in range(len(A)):
    sum = 0
    odd = 0
    even = 0
    
    for col in range(len(A)):
        B[row][col] = A[row][col]
        sum += A[row][col]
        if (A[row][col] % 2) == 0:
            even += 1
        else:
            odd += 1 
        
    B[row][col+1] = sum
    if even > odd:
        B[row][col+2] = 1
    else:
        B[row][col+2] = 0
    
#print(B)

for i,j in enumerate(B):
    print(i)
    for k,l in enumerate(j):
        print(B[i][k])
        print("\n")
```
