Today we will discuss *for* loop.
* For loop is a count controlled loop.
* Unlike *while* loop, we specifically know how many times we should run the loops.

Let's take a look at the how for-loop works.




This looks very similar to the flow-chart for *while* loop. However, we see that there is another element called *loop update.* In the while loop, we do the update inside the loop block. In case of for loop, loop update happens automatically. Excplicit loop update within the loop block is not required.

Let's look at a program written using while loop first. This is the same one we used in the last lecture.

```python
# Let's say we want to print Hello World! five times

count =0

while count < 5:
    print("Hello World!")
    print(count)
    count += 1   # same as count = count + 1
```

Now le'ts write the same program using a for loop.

```python
for i in range(0,5):
    print(i)
    print("Hello World!")
    print("\n")
```

First, let's understand the range function.

```python
print(range(0,5))
print(range(2,6))
```

```python
print(list(range(0,5)))
print(list(range(2,6)))
```

We can observe that the first item in range got printed, but the last one got left out.



The *for* loop can be very handy for a lot of tasks, other than counting.
For exmaple, let's print out the names of the characters in the drama FRIENDS.

```python
FRIENDS = ["Ross", "Rachel", "Monica", "Phoebe", "Chandler", "Joey"]
for character in FRIENDS:
    print(character)
```

```python
# you can also cycle through a string

name = "Lyon College"

for character in name:
    print(character)
```

```python
name = "Lyon College"
len_name = len(name)
for i in range(len(name)):
    print(i)
    print(len_name - i - 1)
    print(name[len_name - i - 1])
    print("\n")
```

```python
for i in range(0,5):
    print(i)
    print("Hello WOrld!")
    print("\n")
    if i == 2:
        break
```

```python
for i in range(0,5):
    if i == 3:
        continue
    print(i)
    print("Hello WOrld!")
    print("\n")
```

```python
### NESTED FOR LOOPS
for i in range(0,4):
    for j in range(0,3):
        #print("i")
        print(i)
        #print("j")
        print(j)
        print(i * j)
        print("\n")
```
