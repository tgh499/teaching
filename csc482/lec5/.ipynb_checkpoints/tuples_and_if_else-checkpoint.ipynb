What are tuples?
* similar to lists, but they are immutable
* Immutability means, once an object is created, it cannot be changed


```python
num_list = [1,2,3,4]
print(num_list)
```

```python
num_tuple = (1,2,3,4)
print(num_tuple)
print(type(num_tuple))
```

* tuples use '(' instead of a '['
* items are separated by ,

Now let's see a special case.


```python
num_list = [1]
print(num_list)
```

```python
num_tuple = (1)
print(num_tuple)
print(type(num_tuple))
```

```python
num_tuple = (1,)
print(num_tuple)
print(type(num_tuple))
```

* since tuples are immutable, they cannot be updated
* you could use list.append() to add items to a list; we cannot do the same for tuples
* however, you can add two tuples and create a new tuple
* you can also do repition, and check for membership
* you can also find the length of a tuple

```python
tup_1 = (1,2,3)
tup_2 = (2,3,4)
new_tuple = tup_1 + tup_2
print(new_tuple)
```

```python
del(tup_1)
print(tup_1)
```

```python
tup_str = ("Lyon ")
new_tuple = tup_str * 4  # repition
print(new_tuple)
```

```python
print(4 in tup_2) # check membership
```

```python
print(len(tup_2)) # check length
```

Now let's discuss control-flow; In real life, we often come across scenarios like - if it rains, take the umbrella, if it doesn't, then don't. Similarly, if I have money, I will buy a ticket to Hawai, otherwise I will not. Many computation problem can also use this if-else structure.

```python
price_of_car = int(input("What's the price of the car?\n"))
if price_of_car > 100000:
    print("OMG!! Out of my budget!")
else:
    print("Let's buy it!!")
```

```python
# program that finds out which number is greater, or if they are equal
num_1 = int(input("Enter a number: "))
num_2 = int(input("Enter another number: "))

if num_1 > num_2:
    print("num_1 is greater")
elif num_1 == num_2:
    print("num_1 and num_2 are equal")
else:
    print("num_2 is greater");
```

Three keywords:
* if
* elif
* else

*if* and *elif* need conditions to work. *else* doesn't need any condition.

```python
# Let's look at one more example

num = 9


if (num >= 10):
    if (num <= 20):
        print("The number is between 10 & 20")
    else:
        print("The number is greater than 20")
else:
    print("The number is greater less than 10")

        
```

```python
# which one will print
num = 4

if num>1:
    print("Hello World!")
elif num>2:
    print("Hello Universe!")
else:
    print("Please try a humber greater than 2")
```
