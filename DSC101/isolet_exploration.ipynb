```python
import pandas as pd
import numpy as np
data = pd.read_csv('isolet5.data', sep=",", header=None)
print(data.shape)
```

```python
y_col = data.columns[617]
x_col = data.columns[0:617]
```

```python
x_test = data[x_col]
y_test = data[y_col]
x_test.head()
```

```python
y_test = y_test.astype(int)
#y_test = np.transpose(y_test)
print(y_test)
```

```python
print(np.min(x_test.values))
```

```python
x_test = x_test - np.min(x_test.values)
```
